```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
data = pd.read_csv('Data-Week6.csv')
data.head()
```

```python
X = data[['FFMC' ,'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind']]
y = data[['month']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print(accuracy_score(y_true=y_test, y_pred=clf.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 20))
M = DecisionTreeClassifier().fit(X,y)
plot_tree(M,filled=True, fontsize=7)
plt.show()
```
