```python
import pandas as pd
import matplotlib.pyplot as plt
headers = ['preg','plas','pres','skin','test','mass','pedi','age','class']
data = pd.read_csv('Data_Week3.csv',names=headers)
data.head(3)
```

Statistics of Data

```python
from pandas import set_option
set_option('display.width', 150)
print(data.shape)
print(data.describe())
```

Class Distribution

```python
count_class = data.groupby('class').size()
count_class
```

Correlation

```python
correl = data.corr(method='pearson')
correl
```

Skewness of Data

```python
data.skew()
```

Data Visualization

```python
data.hist()
plt.show()
```

```python
data.plot(kind='density',subplots=True, layout=(3,3),sharex=False)
plt.show()
```

```python
data.plot(kind='box',subplots=True,layout=(3,3),sharey=False)
plt.show()
```
