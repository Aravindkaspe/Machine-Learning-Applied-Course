Read Iris Data

```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df.head()
```

Read a CSV file

```python
data = pd.read_csv("Data-Week2.csv")
data.head()
```

Visualization - Bar Chart

```python
import matplotlib.pyplot as plt
x = data["County"]
y = data["Married"]
plt.bar(x,y)
plt.xlabel("County")
plt.ylabel("Married")
plt.title("Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data["County"]
y = data["Married"]
plt.pie(y, shadow = False , startangle=90)
plt.xlabel("County")
plt.ylabel("Married")
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data["County"]
y = data["Married"]
plt.plot(x, y)
plt.xlabel("County")
plt.ylabel("Married")
plt.title('Line Chart')
plt.show()
```
