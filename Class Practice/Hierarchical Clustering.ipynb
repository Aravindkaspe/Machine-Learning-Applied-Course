```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Data-Week7.csv")
data.head()
```

```python
X=data[['sqft','price']].values

ID = range(1,20)
plt.figure(figsize=(10,7))
plt.subplots_adjust(bottom=0.1)
# insert the actual value in the frame
plt.scatter(X[:,0],X[:,1], label='True Position')
# insert the ID of each item in the left up of data
for label, x, y in zip(ID, X[:,0], X[:,1]):
    plt.annotate(label, xy=(x,y), xytext=(-3,3), textcoords='offset points',
                ha='right', va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

links = linkage(X, 'single')
plt.figure(figsize=(10,7))
dendrogram(links)
plt.show()
```
