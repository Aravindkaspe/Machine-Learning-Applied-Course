```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart.csv')
data.head(3)
```

```python
selected_attributes = data[['age', 'sex', 'trestbps']]
```

```python
sex_counts = selected_attributes['sex'].value_counts()
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart for Sex')
plt.show()
```

```python
vessels_colored = data['ca']  # Number of Major Vessels Colored by Fluoroscopy
thalassemia = data['thal'] 
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
vessels_colored.value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Major Vessels Colored by Fluoroscopy')
plt.xlabel('Number of Vessels Colored')
plt.ylabel('Count')
plt.show()
```

```python
x=data["cp"].value_counts()
plt.pie(x, autopct='%1.1f%%', labels = ["Typical Angina" , "Atypical Angina", "Non-anginal Pain", "Asymptomatic"], startangle=90)
plt.show()
```
