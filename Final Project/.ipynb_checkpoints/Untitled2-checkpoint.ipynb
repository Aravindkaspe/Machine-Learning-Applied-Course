```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
```

```python
# Load data into a DataFrame
data = pd.read_csv('heart.csv')

# Target variable
target = data["exang"]

print(data.columns)

# Select the two attributes of interest
X = data[["age", "cp"]]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)
```

```python
# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Plot the linear regression
plt.scatter(X_train["age"], y_train, color='blue', label='Actual data')
plt.plot(X_train["age"], model.predict(X_train), color='red', label='Linear regression')
plt.xlabel('Age')
plt.ylabel('Chest Pain')
plt.title('Linear Regression: Age vs. Chest Pain')
plt.legend()
plt.show()
```

```python
# Predict on the test set
predictions = model.predict(X_test).round()

# Create confusion matrix
cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(cm)
```

```python
TN = 112
FP = 31
FN = 19
TP = 43

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", round(accuracy,2))
print("Precision:", round(precision,2))
print("Recall (Sensitivity):", round(recall,2))
print("F1-score:", round(f1_score,2))
```
