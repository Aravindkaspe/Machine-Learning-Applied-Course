```python
from numpy import set_printoptions  
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 
```

```python
X = data.iloc[:,:12] 
Y = data.iloc[:,0]
```

```python
test = SelectKBest(score_func=chi2, k=4) 
fit = test.fit(X,Y)
set_printoptions(precision=2) 
print(fit.scores_) 
featured_data = fit.transform(X) 
print("\nFeatured data:\n", featured_data[0:4])
```

```python
selected_indices = fit.get_support(indices=True)

# Get the feature names
feature_names = X.columns[selected_indices]

# Print feature scores
print("Feature scores:")
print(fit.scores_)

# Print selected feature names
print("\nSelected feature names:")
print(feature_names)

# Transform the data to include only selected features
featured_data = fit.transform(X) 
print("\nFeatured data:\n", featured_data[0:4])
```

```python
from sklearn.feature_selection import RFE 
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression() 
rfe = RFE(model, n_features_to_select=4) 
fit = rfe.fit(X, Y) 
print(fit.support_)
print(fit.ranking_)
selected_indices = fit.get_support(indices=True)

# Get the feature names
feature_names = X.columns[selected_indices]

# Print selected feature names
print("\nSelected feature names:")
print(feature_names)
```

```python
from sklearn.ensemble import ExtraTreesClassifier 
model = ExtraTreesClassifier() 
model.fit(X, Y) 
feature_importances = model.feature_importances_

# Create a DataFrame to store feature names and their importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort features by their importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importances
print("Feature importances:")
print(feature_importance_df)

# Get the selected features based on a threshold or a fixed number
threshold = 0.1  # Adjust this threshold as needed
selected_features = feature_importance_df[feature_importance_df['Importance'] >= threshold]['Feature'].tolist()

# Print the selected features
print("\nSelected features:")
print(selected_features)
print(model.feature_importances_)
```
